```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine("sqlite:///../superstore.db")
```

```python
df = pd.read_sql("SELECT * FROM superstore", engine)
```

```python
query = """
SELECT "Product Name",
       SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY "Product Name"
ORDER BY Total_Sales DESC
LIMIT 5;
"""

top_products = pd.read_sql(query, engine)
top_products
```

```python
query = """
SELECT Category,
       SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY Category;
"""

category_sales = pd.read_sql(query, engine)
category_sales
```

```python
query = """
SELECT "Customer Name",
       SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY "Customer Name"
ORDER BY Total_Sales DESC
LIMIT 10;
"""

top_customers = pd.read_sql(query, engine)
top_customers
```

```python
query = """
SELECT Region,
       SUM(Profit) AS Total_Profit
FROM superstore
GROUP BY Region;
"""

region_profit = pd.read_sql(query, engine)
region_profit
```

```python
query = """
SELECT "Order Date", Sales
FROM superstore;
"""

monthly_sales = pd.read_sql(query, engine)
monthly_sales.head()
```
