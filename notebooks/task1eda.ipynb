```python
import pandas as pd

df = pd.read_csv("../data/Sample - Superstore.csv", encoding="latin1")
Q1 = df["Sales"].quantile(0.25)
Q3 = df["Sales"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df["Sales"] < lower) | (df["Sales"] > upper)]

print("Number of Outliers:", len(outliers))
```

```python
import pandas as pd

df = pd.read_csv("../data/Sample - Superstore.csv", encoding="latin1")
Q1 = df["Profit"].quantile(0.25)
Q3 = df["Profit"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

profit_outliers = df[(df["Profit"] < lower) | (df["Profit"] > upper)]

print("Number of Profit Outliers:", len(profit_outliers))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/Cleaned_Superstore.csv")
df.info()
df.describe()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Sales"], bins=30)
plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df["Sales"])
plt.title("Boxplot of Sales")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Category", data=df)
plt.title("Number of Orders by Category")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
sales = df.groupby("Order Date")["Sales"].sum()

plt.figure(figsize=(12,5))
plt.plot(sales)
plt.title("Sales Trend Over Time")
plt.xlabel("Order Date")
plt.ylabel("Sales")
plt.show()
```

```python
top_states = df.groupby("State")["Sales"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
top_states.plot(kind="bar")
plt.title("Top 10 States by Sales")
plt.xlabel("State")
plt.ylabel("Sales")
plt.show()
```

# Key Insights

1. Technology category generated the highest sales.

2. Furniture category had comparatively lower profit.

3. Some transactions recorded negative profit.

4. Sales show fluctuations over time.

5. A few states contribute significantly more sales than others.
